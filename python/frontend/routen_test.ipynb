```python
import requests

url = "http://localhost:8000"

response_get_root = requests.get(url) # Request getätigt und die Response wird in die variable "respose" gespeichert
print(response_get_root.json())

payload = {
    "name": "Chucky",
    "email": "gmail@chucknorris.com",
    "password": "123457",
    "marketing_opt_in": True
    }

response_post_new_user = requests.post(url + "/new_user", json=payload)

print(response_post_new_user.json())
```
